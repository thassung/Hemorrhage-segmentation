```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'Dataset\BHSD_original\images\ID_0b10cbee_ID_f91d6a7cd2.nii.gz'
```

```python
img_obj = nib.load(path)
img_data = img_obj.get_fdata()
plt.imshow(img_data[:, :, 16], cmap='gray')
plt.show()
```

```python
print(img_data[:, :, 16])
```

# Windowing

```python
img_obj = nib.load(path)
img_data = img_obj.get_fdata()

slope = img_obj.dataobj.slope
intercept = img_obj.dataobj.inter

img_data[img_data >= 50] = 50
img_data[img_data <= 20] = 20
images = slope * img_data.astype(np.float32)
hu_images = images + intercept
```

```python
plt.imshow(hu_images[:, :, 16], cmap='gray')
plt.show()
```
